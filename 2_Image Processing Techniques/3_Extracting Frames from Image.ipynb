```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

## Load Image 

```python
img = cv.imread('transition.jpg')

img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
img.shape
```

## Slicing the Image

```python
plt.imshow(img[:, :500, : ])
```

```python
plt.imshow(img[ :1000, : , : ])
```

## Extracting Frames

```python
r = img[:,:,0]   # -> Red

g = img[:,:,1]  # -> Green

b = img[:,:,2]   # -> Blue
```

```python
plt.imshow(r)
cv.imwrite('red_t.png',r)
```

```python
plt.imshow(g)
cv.imwrite('green_t.png',g)
```

```python
plt.imshow(b)
cv.imwrite('blue_t.png',b)
```
