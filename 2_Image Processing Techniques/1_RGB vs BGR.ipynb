```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

## Import image using openCV

```python
img = cv.imread('rgb.jpeg')

plt.imshow(img)
```

```python
# RGB <-> BGR
# BGR <-> RGB
```

## Convert BGR to RGB

```python
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img2 = cv.imread('transition.jpg')
plt.imshow(img2)
```

```python
img2 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
plt.imshow(img2)
```
