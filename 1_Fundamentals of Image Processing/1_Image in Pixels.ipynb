## Image at Pixel Level

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
arr = np.array([[0,0,0],
                [0,1,0],
                [0,0,0]])
```

```python
arr
```

```python
plt.imshow(arr)
```

```python
chess = np.zeros((8,8),dtype=int)

for i in range(8):
    for j in range(8):
        if ((i+j) % 2 == 0):
            chess[i][j] = 1
plt.imshow(chess)
```

## High Pixel Density Images

```python
cv.imwrite('img.png',chess)
```

```python
lst = []
for i in range(720):
    temp = []
    for j in range(1280):
        temp.append(255)
    lst.append(temp)
lst = np.array(lst)

plt.imshow(lst)
```

```python
cv.imwrite('white.png',lst)
```

```python
shades = []
for i in range(256):
    temp = [] 
    for j in range(256):
        temp.append(j)
    shades.append(temp)

shades = np.array(shades)
```

```python
cv.imwrite('shades.png',shades)
```
