```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

## Create Custom Colors

```python
arr = np.array([[[ 249  , 220 , 104 ] , [ 55  , 126 , 34 ] ,[ 90  , 0 , 0 ]],
                [[ 139  , 206 , 247 ] , [ 239 , 134 , 59 ] ,[ 180 , 0 , 0 ]],
                [[ 113  , 178 , 229 ] , [ 240 , 0 , 0 ] ,[ 255 , 0 , 0 ]]])

plt.imshow(arr)
```

```python
orange = [] 
for i in range(720):
    temp = []
    for j in range(1280):
        temp.append([239, 134,59])
    orange.append(temp)
orange = np.array(orange)
plt.imshow(orange)
cv.imwrite('orange.png',orange)
```

```python
img = []

for i in range(720):
    t = []
    for j in range(1080):
        t.append([ 59 , 134 , 239])
    img.append(t)

img = np.array(img)

plt.imshow(img)

cv.imwrite('Orange_1.png',img)
```

```python
lst = []
for i in range(256):
    temp = []
    for j in range(1280):
        temp.append(256+i)
    lst.append(temp)
lst = np.array(lst)
plt.imshow(lst)
```
