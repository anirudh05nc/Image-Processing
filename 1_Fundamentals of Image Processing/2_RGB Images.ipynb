## Creating Dense images

```python
# RED DENSE IMAGE
red_dense = []
for i in range(256):
    temp = []
    for j in range(256):
        temp.append([i, 0, 0])
    red_dense.append(temp)

red_dense = np.array(red_dense)
plt.imshow(red_dense)
```

```python
# GREEN DENSE IMAGE
green_dense = []
for i in range(256):
    temp = []
    for j in range(256):
        temp.append([0, i, 0])
    green_dense.append(temp)
green_dense = np.array(green_dense)
plt.imshow(green_dense)
```

```python
# BLUE DENSE IMAGE
blue_dense = []
for i in range(256):
    temp = []
    for j in range(256):
        temp.append([0, 0, i])
    blue_dense.append(temp)
blue_dense = np.array(blue_dense)
plt.imshow(blue_dense)
```

## Merging Dense Images

```python
arr = []

for i in range(256):
    temp = []
    for j in range(1536):
        temp.append([i, 0, 0])     #Black to Red
    arr.append(temp)
for i in range(256,-1,-1):
    temp = []
    for j in range(1536):
        temp.append([i, 0, 0])    #Red to Black
    arr.append(temp)

for i in range(256):
    temp = []
    for j in range(1536):
        temp.append([0, i, 0])     #Black to Green
    arr.append(temp)
for i in range(256,-1,-1):
    temp = []
    for j in range(1536):
        temp.append([0, i, 0])    #Green to Black
    arr.append(temp)

for i in range(256):
    temp = []
    for j in range(1536):
        temp.append([0, 0, i])     #Black to Blue
    arr.append(temp)
for i in range(256,-1,-1):
    temp = []
    for j in range(1536):
        temp.append([0, 0, i])    #Blue to Black
    arr.append(temp)
plt.imshow(arr)
arr = np.array(arr)
cv.imwrite('transition.jpg',arr)
```
